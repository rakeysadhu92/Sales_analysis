# Importing Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
all_data = pd.DataFrame()
files = os.listdir('Sales_Data')
files
```

```python
for file in files:
    sample = pd.read_csv('Sales_Data/'+file)
    all_data = pd.concat([all_data, sample])
all_data.sample(5)
```

###Best Month for sales and How much was earned that Month?.

```python
#Adding a month to every row.
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data.drop(all_data.loc[all_data['Order ID']=='Order ID'].index, inplace=True)

#Finding the NaN values in the dataframe
nan_df = all_data[all_data.isna().any(axis=1)]

#Droping the NaN values in the dataframes
all_data = all_data.dropna(how='all')

all_data.head()
```

```python
# Converting the Month Column to an integer values

all_data['Month'] = all_data['Month'].astype('int32')
all_data.sample(5)
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
all_data.sample(10)
```

```python
all_data['Sales in Dollars'] = all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

```python
monthly_sales = all_data.groupby('Month').sum()
monthly_sales.head()
```

```python
# plt.plot(monthly_sales['Month'], monthly_sales['Sales in Dollars'], kind='bar')
# plt.show()

monthly_sales.plot.bar(y = 'Sales in Dollars')
plt.xlabel('Month')
plt.ylabel('Sales in Dollars')
```

### What Products are Sold Together?.

```python
all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df.head()
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for x in df['Grouped']:
    y = x.split(',')
    count.update(Counter(combinations(y,2)))

for key, value in count.most_common(15):
    print(key, value)
```
